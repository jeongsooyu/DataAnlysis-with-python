```python
title=[]
rate=[]
address=[]
food=[]
hit=[]
review=[]
favorite=[]
```

```python
for i in tqdm(range(50)):
    href = href_list[i]
    html = urlopen(href)
    soup = BeautifulSoup(html, 'html.parser')

    title.append(soup.select('span.title > h1')[0].get_text().strip())
    rate.append(float(soup.select('strong.rate-point')[0].get_text().strip()))
    address.append(soup.select('ul > li > div')[1].get_text().strip())
    food.append(soup.select('tbody > tr > td > span')[0].get_text())
    hit.append(int(soup.select('header span')[3].get_text().replace(',','')))
    review.append(int(soup.select('header span')[4].get_text().replace(',','')))
    favorite.append(int(soup.select('header span')[5].get_text().replace(',','')))
```

```python
#지금까지 얻은 데이터를 모두 DataFrame으로 저장
data = {'Rank' : [i+1 for i in range(50)],
        'Name' : title,
        'Food' : food,
        'Rate' : rate,
        'Hit' : hit,
        'Review' : review,
        'Favorite' : favorite,
        'Address' : address,
        'URL' : href_list}
df = pd.DataFrame(data)
df.head()
```
