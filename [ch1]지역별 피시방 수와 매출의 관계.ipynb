```python
PC_sales = pd.read_excel('data/ch01.PC_Room_area_and_sales.xls', encoding='utf-8')
PC_sales
```

```python
df_sales=pd.DataFrame(PC_sales)
```

```python
df_sales=df_sales.drop(['통계분류(1)'], axis=1)
```

```python
df_sales
```

```python
df_sales.rename({df_sales.columns[1] : '2012 총합',
                 df_sales.columns[2] : '2012 PC방 매출',
                 df_sales.columns[3] : '2013 총합',
                 df_sales.columns[4] : '2013 PC방 매출',
                 df_sales.columns[5] : '2014 총합',
                 df_sales.columns[6] : '2014 PC방 매출',
                 df_sales.columns[7] : '2015 총합',
                 df_sales.columns[8] : '2015 PC방 매출',
                 df_sales.columns[9] : '2016 총합',
                 df_sales.columns[10] : '2016 PC방 매출'}, axis=1, inplace=True)
```

```python
pc_list=[name for name in df_sales.columns if 'PC' in name]
pc_list.insert(0,'구별')
```
